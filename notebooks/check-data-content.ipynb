```python
import os
import scipy
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
```

```python
data_path = '../data/sample.xlsx'

# load excel sheet
sleepy_df = pd.read_excel(data_path, sheet_name=0)
unsleepy_df = pd.read_excel(data_path, sheet_name=1)
print('number of samples(sleepy) -> {}'.format(len(sleepy_df)))
print('number of samples(not sleepy) -> {}'.format(len(sleepy_df)))

# add new column that indicate category
sleepy_df['sleepy'] = 1
unsleepy_df['sleepy'] = 0

# cancatenate two dataframe so that you can manupulate data easily
concatnated_df = pd.concat([sleepy_df, unsleepy_df])

# plot distribution of RRI value
sleepy_RRIs = concatnated_df.query('sleepy == 1')['RRI']
unsleepy_RRIs = concatnated_df.query('sleepy == 0')['RRI']
plt.figure(figsize=(8,4))
plt.boxplot([sleepy_RRIs, unsleepy_RRIs], labels=['sleepy', 'unsleepy'])
plt.title('RRI')
plt.show()

# plot distribution of temperature value
sleepy_RRIs = concatnated_df.query('sleepy == 1')['temperature']
unsleepy_RRIs = concatnated_df.query('sleepy == 0')['temperature']
plt.figure(figsize=(8,4))
plt.boxplot([sleepy_RRIs, unsleepy_RRIs], labels=['sleepy', 'unsleepy'])
plt.title('Temperature')
plt.show()

# plot distribution of x-axis acceleration
sleepy_RRIs = concatnated_df.query('sleepy == 1')['x']
unsleepy_RRIs = concatnated_df.query('sleepy == 0')['x']
plt.figure(figsize=(8,4))
plt.boxplot([sleepy_RRIs, unsleepy_RRIs], labels=['sleepy', 'unsleepy'])
plt.title('X-axis acceleration')
plt.show()

# plot distribution of y-axis acceleration
sleepy_RRIs = concatnated_df.query('sleepy == 1')['y']
unsleepy_RRIs = concatnated_df.query('sleepy == 0')['y']
plt.figure(figsize=(8,4))
plt.boxplot([sleepy_RRIs, unsleepy_RRIs], labels=['sleepy', 'unsleepy'])
plt.title('Y-axis acceleration')
plt.show()

# plot distribution of z-axis acceleration
sleepy_RRIs = concatnated_df.query('sleepy == 1')['z']
unsleepy_RRIs = concatnated_df.query('sleepy == 0')['z']
plt.figure(figsize=(8,4))
plt.boxplot([sleepy_RRIs, unsleepy_RRIs], labels=['sleepy', 'unsleepy'])
plt.title('Z-axis acceleration')
plt.show()
```
